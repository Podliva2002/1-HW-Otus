```python
from bs4 import BeautifulSoup

import requests
```

```python
def get_https_links(url, visited, timeout=5, limit=2000, count=[0]):
    if url in visited or count[0] >= limit:
        return
    
    visited.add(url)

    try:
        response = requests.get(url, timeout=timeout)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        https_links = soup.find_all('a', href=lambda href: href and 'https' in href)

        links = [link['href'] for link in https_links]

        for link in links:
            if count[0] < limit:
                print(link)
                count[0] += 1
            else:
                break

        for link in links:
            if count[0] < limit:
                get_https_links(link, visited, timeout, limit, count)

    except requests.RequestException as e:
        print(f"Ошибка при получении страницы {url}: {e}")

start_url = 'https://habr.com/ru/articles/51772/'
visited_urls = set()
get_https_links(start_url, visited_urls)
```
